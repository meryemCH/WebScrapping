```python
#import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#saisir l'url
url = 'https://www.tripadvisor.fr/Hotels-g187147-Paris_Ile_de_France-Hotels.html'
```

```python
#import le code de la page
reponse = requests.get(url)
soup = BeautifulSoup(reponse.text,"html.parser")
soup
```

```python
#les info sont dans les divs
divs=soup.findAll('div')
```

```python
#affichage
divs
```

```python
title= soup.findAll('div',{'class':'listing_title'})
title
```

```python
price= soup.find_all('div', {'class': 'price-wrap'})
price
```

```python
description=soup.find_all('span',{'class':'text_container'})
description
```

```python
reviews=soup.find_all('a',{'class':'review_count'})
reviews
```

```python
photos=soup.find_all('img',{'class':'cxohC _C h w'})
photos
```

```python
news_divs=[]
```

```python
for i in divs :
  news_i={}
  news_i['title']=i.title.text
  news_i['price']=i.title.text
  news_i['description']=i.title.text
  news_i['reviews']=i.title.text
  news_i['photos']=i.title.text
  news_divs.append(news_i)
```

```python
news_divs
```

```python
#convertir la liste en base de données 
df= pd.DataFrame(divs,columns=['title','price','description','reviews','photos'])
```
