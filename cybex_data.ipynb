```python
#import packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#saisir l'url
url = 'https://cybex-online.com/fr-fr/car-seats?line=Platinum'
```

```python
#import le code de la page
reponse = requests.get(url)
soup = BeautifulSoup(reponse.text,"html.parser")
soup
```

```python
product= soup.findAll('a',{'class':"Nav-item u-sm-hidden is-active"})
product
```

```python
price= soup.find_all('div', {'class': 'price-wrap'})
price
```

```python
description=soup.find_all('span',{'class':'text_container'})
description
```

```python
reviews=soup.find_all('a',{'class':'review_count'})
reviews
```

```python
photos=soup.find_all('img',{'class':'cxohC _C h w'})
photos
```
